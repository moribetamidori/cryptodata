```python
import numpy as np
import pandas as pd
# import seaborn as sns
import json
from glom import  glom

with open('data.json','r') as f:
    data = json.loads(f.read())
df = pd.json_normalize(data, record_path=['data'])
new_df = df[['name','symbol','max_supply','circulating_supply','total_supply','quote.USD.market_cap','quote.USD.market_cap_dominance','quote.USD.fully_diluted_market_cap','quote.USD.last_updated']]
new_df
# new_df['quote.USD.market_cap_dominance'].sum()
```

```python
#Smart contract platforms / Ecosystems token 
eth = new_df.loc[new_df['symbol']=='BTC']
sol = new_df.loc[new_df['symbol']=='SOL']
ada = new_df.loc[new_df['symbol']=='ADA']
avax = new_df.loc[new_df['symbol']=='AVAX']
matic = new_df.loc[new_df['symbol']=='MATIC']
smart_contract_token = pd.concat([eth, sol, ada, avax, matic], axis=0)
smart_contract_token
ax = smart_contract_token.plot.bar(x='symbol', y = 'quote.USD.market_cap_dominance')
```

```python
#NFT token 
axs = new_df.loc[new_df['symbol']=='AXS']
theta = new_df.loc[new_df['symbol']=='THETA']
sand = new_df.loc[new_df['symbol']=='SAND']
mana = new_df.loc[new_df['symbol']=='MANA']
gala = new_df.loc[new_df['symbol']=='GALA']
nft_token = pd.concat([axs, theta, sand, mana, gala], axis=0)
nft_token
```

```python
#Meme token 
doge = new_df.loc[new_df['symbol']=='DOGE']
shib = new_df.loc[new_df['symbol']=='SHIB']
mim = new_df.loc[new_df['symbol']=='MIM']
meme_token = pd.concat([doge, shib, mim], axis=0)
meme_token
```

```python
#DeFi token 
luna = new_df.loc[new_df['symbol']=='LUNA']
link = new_df.loc[new_df['symbol']=='LINK']
uni = new_df.loc[new_df['symbol']=='UNI']
dai = new_df.loc[new_df['symbol']=='DAI']
steth = new_df.loc[new_df['symbol']=='STETH']
defi_token = pd.concat([luna, link, uni, dai, steth], axis=0)
defi_token 
# need to change ceth to steth, anyways, luna has a dual coin system,n
# Terra is the price-stable coin which users can use for payment purposes.
# Luna is the coin working in the background and held by the stakeholders.
# Luna’s primary utility is to protect Terra from price fluctuations.
# they both share the same symbol? How about Uniswap? 
```
